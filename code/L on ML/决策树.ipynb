# 决策树
## 树模型
- 从根节点开始一步步走到叶子节点（决策）
- 所有的数据最终都会落到叶子节点（最后的位置），<mark>既可以做分类也可以做回归</mark>
### 树的组成
- 根节点：第一个选择点
- 非叶子节点与分支：中间过程
- 叶子节点：最终的决策结果


### 决策树的训练与测试
- 训练阶段：从给定的训练集构造出来一棵树（从根节点开始选择特征，如何进行特征切分）
- 测试阶段：根据构造出来的树模型从上到下走一遍就好了
### 如何切分特性
- 目标：通过一个衡量标准，来计算通过不同特征进行分之选择后的分类情况，找出来最好的当成根节点
- 衡量标准：熵 —— 通过节点分枝将数据的熵值下降
  $ H(x) = - \sum p_i * \log p_i, i = 1,2,...,n $
- 信息增益：表示特征X使得类Y的不确定性减少的程度 $\sum H(x_i) < H(X)$
